** Plotting USA normalized values **

We thought that heatmaps would be the best possible way to represent a huge number of results. As you can see in the resulting heatmaps it is quite easy to interpret the results in the visualisations, as well as make comparisons among different states.

For our results, we decided to look a four industries in particular:
* Technology
* Finance 
* Health Care
* Media and Entertainment

The reasons for choosing the above for industries is because not only are they extremely topical, but Technology, Health Care, and Media and Entertainment were the top 3 industries in the dataset. Finance wasn't far behind, but we chose finance because it is a very relatable and prominent industry.


** Technology **

```python
state_geo = os.path.join('../../data/folium_files/us-states.json')
technology_heatmap = folium.Map(location=[37, -102], zoom_start=4)
technology_heatmap.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=normalized_usa,
 columns=['state_code', 'Technology'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.3,
 legend_name='Normalized funding fraction'
)
folium.LayerControl().add_to(technology_heatmap)

technology_heatmap
```

It is very interesting looking at the above graph of the normalized values for technology. Before looking at the map one may be inclined to assume that the 'best' place for setting up a Technology company in terms of the impact it has on funding would be California. Although as we can see from the above map this is not neccessarily true. Idaho (ID) is the best place to set up a technology company according to the our results. This is most likely due to the low number of companies in other industries in Idaho, as well as Technology dominating the funding in Idaho. Overall we found this result to be very surprising. 

** Finance **

```python
state_geo = os.path.join('../../data/folium_files/us-states.json')
finance_heatmap = folium.Map(location=[37, -102], zoom_start=4)

finance_heatmap.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=normalized_usa,
 columns=['state_code', 'Finance'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.3,
 legend_name='Normalized funding fraction'
)
folium.LayerControl().add_to(finance_heatmap)

finance_heatmap
```

It is often thought that NYC is the hub of finance, but according to our results, NYC is not the best place to setup a finance company in terms of the impact it will have on your chances of getting funded. Georgia and Florida seem to be the two front runners in this case. Once again these were quite unexpected results.
